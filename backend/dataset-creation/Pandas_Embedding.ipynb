#### Crating Huggingface Dataset

```python
from datasets import Dataset

hf_dataset = Dataset.from_pandas(df)
```

```python
hf_dataset
```

```python
from transformers import AutoTokenizer, AutoModel

checkpoint = "sentence-transformers/multi-qa-mpnet-base-dot-v1"       # produces 768 dimensional embedding

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModel.from_pretrained(checkpoint)
```

```python
# Using CLS pooling for embedding generation

def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]    # embedding of [CLS] token
```

```python
def get_embeddings(text_list):

    encoded_input = tokenizer(
        text_list,
        padding=True,
        truncation=True,
        return_tensors="pt"
    )

    model_output = model(**encoded_input)

    return cls_pooling(model_output)
```

```python
# generating embedding for a example

emb = get_embeddings(hf_dataset[0]['content'])
```

```python
emb.shape
```

```python
embedding_dataset = hf_dataset.map(
    lambda x: {"embeddings": get_embeddings(x["content"]).detach().cpu().numpy()[0]}
)
```
